если будем использовать API (*я планирую просто скачать всю базу, но я пока что с мобильного*, а API - для обновления и тестирования)

```
#@title Код получения информации о преступлениях через API { display-mode: "form" }


!pip install sodapy
import pandas as pd
from sodapy import Socrata

#TO DO: crimes_df = pd.DataFrame.from_csv("Crimes_-_2001_to_present.csv")

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cityofchicago.org", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.cityofchicago.org,
#                  MyAppToken,
#                  userame="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
crimes = client.get("6zsd-86xi", limit=2000)

# Convert to pandas DataFrame
crimes_df = pd.DataFrame.from_records(crimes)

```

Покажем на карте полученные точки. Для удобства также возьмём [границы города](https://dev.socrata.com/foundry/data.cityofchicago.org/qqq8-j68g)

```
#@title Код получения информации о границе города через API а также отображения карты { display-mode: "form" }


# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cityofchicago.org", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.cityofchicago.org,
#                  MyAppToken,
#                  userame="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
city = client.get("qqq8-j68g", limit=2000)

# Convert to pandas DataFrame
city_MultyPoligon = pd.DataFrame.from_records(city)['the_geom'][0]['coordinates']

#!apt-get install libgeos-3.5.0
#!apt-get install libgeos-dev
#!pip install https://github.com/matplotlib/basemap/archive/master.zip
!pip install pyproj==1.9.6
from mpl_toolkits.basemap import Basemap
from matplotlib import pyplot as plt
def map_of_chicago():
  fig = plt.figure(figsize=(8, 8))
  m = Basemap(projection='lcc', resolution='h', 
            lat_0=41.8, lon_0=-87.7,
            width=5E4, height=5.5E4)
  m.shadedrelief()
  m.drawcoastlines(color='gray')
  m.drawcountries(color='gray')
  m.drawstates(color='gray')

  for poly in city_MultyPoligon:
    for lines in poly:
      m.plot([float(x[0]) for x in lines],
          [float(x[1]) for x in lines],
           'g--',
           latlon=True)
  return m     

map_of_chicago()
```

```
#@title Отрисовка карты преступлений { display-mode: "form" }
list(crimes_df)
crimes_df_located = crimes_df[pd.notnull(crimes_df['longitude'])]
map_of_chicago().scatter(list(crimes_df_located['longitude'].astype(float)),
          list(crimes_df_located['latitude'].astype(float)),
         latlon=True)
```
